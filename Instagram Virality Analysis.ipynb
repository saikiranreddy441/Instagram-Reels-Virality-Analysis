```python
# Convert Boolean columns to 1 and 0
df['trending_audio'] = df['trending_audio'].astype(int)
df['explore_page_boost'] = df['explore_page_boost'].astype(int)
```

```python
# Create Original Virality Category using fixed score boundaries

df['original_virality_category'] = np.select(
    [
        df['virality_score'] <= 25,
        df['virality_score'] <= 50,
        df['virality_score'] <= 75
    ],
    [
        'Low',
        'Medium',
        'High'
    ],
    default='Viral'
)
```

```python
# Calculate 90th and 10th percentile thresholds
top_10_threshold = df['virality_score'].quantile(0.90)
bottom_10_threshold = df['virality_score'].quantile(0.10)

print(f"Top 10% Threshold: {top_10_threshold:.2f}")
print(f"Bottom 10% Threshold: {bottom_10_threshold:.2f}")
```

```python
# Create Original Virality Group
df['original_virality_group'] = np.select(
    [
        df['virality_score'] >= top_10_threshold,
        df['virality_score'] <= bottom_10_threshold
    ],
    [
        'Top 10%',
        'Bottom 10%'
    ],
    default='Middle 80%'
)
```

```python
df['original_virality_group'].value_counts()
```

```python
df[['virality_score', 'original_virality_group']].head(10)
```

```python
# Columns used for Business Virality Index
bvi_columns = [
    'non_follower_reach_ratio',
    'shares',
    'engagement_rate',
    'saves',
    'reach',
    'impressions'
]

# Min-Max Normalization
for col in bvi_columns:
    df[col + '_norm'] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())
```

```python
df[[col + '_norm' for col in bvi_columns]].head()
```

```python
for col in bvi_columns:
    print(f"{col}_norm")
    print("Min:", df[col + '_norm'].min())
    print("Max:", df[col + '_norm'].max())
    print("-"*30)
```

```python
# Calculate Business Virality Index (BVI)

df['business_virality_index'] = (
    0.30 * df['non_follower_reach_ratio_norm'] +
    0.25 * df['shares_norm'] +
    0.20 * df['engagement_rate_norm'] +
    0.10 * df['saves_norm'] +
    0.10 * df['reach_norm'] +
    0.05 * df['impressions_norm']
) * 100
```

```python
df[[
    'non_follower_reach_ratio_norm',
    'shares_norm',
    'engagement_rate_norm',
    'saves_norm',
    'reach_norm',
    'impressions_norm',
    'business_virality_index'
]].head()
```

```python
df['business_virality_index'].describe()
```

```python
# Calculate Quartile Thresholds for BVI

q1 = df['business_virality_index'].quantile(0.25)
q2 = df['business_virality_index'].quantile(0.50)
q3 = df['business_virality_index'].quantile(0.75)

print(f"Q1 (25th Percentile): {q1:.2f}")
print(f"Q2 (50th Percentile): {q2:.2f}")
print(f"Q3 (75th Percentile): {q3:.2f}")
```

```python
# Create Business Virality Category using Quartiles

df['business_virality_category'] = np.select(
    [
        df['business_virality_index'] <= q1,
        df['business_virality_index'] <= q2,
        df['business_virality_index'] <= q3
    ],
    [
        'Low',
        'Medium',
        'High'
    ],
    default='Viral'
)
```

```python
df[['business_virality_index', 'business_virality_category']].head(10)
```

```python
df['business_virality_category'].value_counts()
```

```python
# Calculate the 90th and 10th percentile thresholds for BVI

bvi_top_10 = df['business_virality_index'].quantile(0.90)
bvi_bottom_10 = df['business_virality_index'].quantile(0.10)

print(f"Top 10% Threshold : {bvi_top_10:.2f}")
print(f"Bottom 10% Threshold : {bvi_bottom_10:.2f}")
```

```python
# Create Business Virality Group

df['business_virality_group'] = np.select(
    [
        df['business_virality_index'] >= bvi_top_10,
        df['business_virality_index'] <= bvi_bottom_10
    ],
    [
        'Top 10%',
        'Bottom 10%'
    ],
    default='Middle 80%'
)
```

```python
df[['business_virality_index', 'business_virality_group']].head(10)
```

```python
df['business_virality_group'].value_counts()
```

```python
# Store all important thresholds

thresholds = {
    "Original Virality Top 10%": top_10_threshold,
    "Original Virality Bottom 10%": bottom_10_threshold,
    "BVI Q1": q1,
    "BVI Median": q2,
    "BVI Q3": q3,
    "BVI Top 10%": bvi_top_10,
    "BVI Bottom 10%": bvi_bottom_10
}

for key, value in thresholds.items():
    print(f"{key}: {value:.2f}")
```
